```python
import xml.etree.ElementTree as ET
import pandas as pd

# Parse the XML file
tree = ET.parse('ProviderFiles/741152597_memorial-hermann-texas-medical-center_standardcharges.xml')

# Get the root element
root = tree.getroot()

# Find the 'Charge Type' element with the attribute 'MS-DRG'
charge_type = root.find(".//Charge[@Type='MS-DRG']")

# Initialize an empty list to hold the data
data = []

# Iterate over each child element in charge_type
for child in charge_type:
    # Initialize an empty dictionary to hold the row data
    row = {}    
    if child.tag == 'Item':
        row['Item Code'] = child.attrib['Code']
    # Iterate over each sub-element in child
    for subchild in child:
        # Add the sub-element tag and text as a key-value pair in the dictionary
        row[subchild.tag] = subchild.text
        # If the sub-element is 'Item', add 'Item Code' to the dictionary

        # If the sub-element is 'Contracts', add a new column for each 'Contract'
        if subchild.tag == 'Contracts':
            for contract in subchild:
                row[contract.attrib['Payer']] = contract.attrib['Charge']
    # Append the dictionary to the list
    data.append(row)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data)
df
```
